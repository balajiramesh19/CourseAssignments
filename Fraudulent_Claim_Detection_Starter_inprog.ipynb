#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
import pandas as pd

def analyse_target_likelihood_for_categoricals(df, target_col, positive_class='Y', threshold=0.05):
  
    df = df.copy()

    # Step 1: Convert target to binary
    df['__target_binary'] = df[target_col].apply(lambda x: 1 if x == positive_class else 0)

    # Step 2: Identify categorical features
    categorical_cols = df.select_dtypes(include='object').columns.tolist()
    if target_col in categorical_cols:
        categorical_cols.remove(target_col)

    likelihood_dict = {}
    uninformative_features = []

    # Step 3: Compute target likelihood for each category in each categorical feature
    for col in categorical_cols:
        try:
            likelihood = df.groupby(col)['__target_binary'].mean()
            if likelihood.empty:
                continue  # Skip if no data
            
            likelihood_df = likelihood.reset_index()
            likelihood_df.rename(columns={'__target_binary': 'target_likelihood'}, inplace=True)
            likelihood_dict[col] = likelihood_df

            # Step 4: Check if variation is below the threshold
            variation = likelihood_df['target_likelihood'].max() - likelihood_df['target_likelihood'].min()
            if variation < threshold:
                uninformative_features.append(col)

        except Exception as e:
            print(f"Error processing column '{col}': {e}")
            continue

    return likelihood_dict, uninformative_features

likelihoods, low_info_cols = analyse_target_likelihood_for_categoricals(df_new, 'fraud_reported')
# likelihoods, low_info_cols = analyse_target_likelihood_for_categoricals(df_new_cleaned, 'fraud_reported')

print("Uninformative categorical features:", low_info_cols)
display(likelihoods)
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
def combine_low_frequency_categories(df, categorical_cols, threshold=0.01):
    df = df.copy()
    total_rows = len(df)

    for col in categorical_cols:
        freq = df[col].value_counts(normalize=True)
        low_freq_categories = freq[freq < threshold].index
        df[col] = df[col].apply(lambda x: 'Other' if x in low_freq_categories else x)

    return df


categorical_cols = X_train.select_dtypes(include='object').columns.tolist()

# Apply to both training and validation sets
X_train = combine_low_frequency_categories(X_train, categorical_cols, threshold=0.05)
X_val = combine_low_frequency_categories(X_val, categorical_cols, threshold=0.05)

print(X_train.shape)
print(X_val.shape)
df.head()
```

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np

# Define ROC function
def plot_roc_curve( actual, predicted, pos_label='Y', title='ROC Curve'):

    # Compute ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(actual, predicted)
    auc_score = roc_auc_score(actual, predicted)

    # Plot ROC curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {auc_score:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.show()

    return auc_score

   
   
```
